```
import matplotlib.pyplot as plt
import numpy as np

# Datos técnicos comparativos (Escala 1-10 de impacto negativo)
# Métricas: [Carga Glucémica, Deshidratación, Inflamación/Congéneres]
cerveza = [6, 5, 5]    # Carga media-alta (maltosa), deshid y inflam moderada
vino = [4, 5, 8]       # Carga baja, deshid moderada, INFLAMACIÓN ALTA (taninos/sulfitos)
destilado_agua = [1, 4, 3] # BAJO EN TODO (Control de Daños Ideal)
destilado_coca_normal = [10, 8, 9] # PEOR ESCENARIO (Azúcar + Alcohol + Cafeína)
destilado_coca_zero = [1, 6, 4] # Mejor que azúcar, peor que agua para deshidratación

# Preparar datos para el gráfico de barras agrupadas
labels = ['Carga Glucémica', 'Deshidratación', 'Inflamación']
categories = ['Cerveza', 'Vino', 'Destilado + Agua', 'Destilado + Coca-Cola', 'Destilado + Coca-Zero']
colors = ['#f1c40f', '#9b59b6', '#3498db', '#e74c3c', '#34495e']
data = [cerveza, vino, destilado_agua, destilado_coca_normal, destilado_coca_zero]

x = np.arange(len(labels))  # Posiciones de las etiquetas
width = 0.15  # Ancho de las barras

fig, ax = plt.subplots(figsize=(14, 7))

# Crear barras para cada categoría
for i, (category_data, category_label, color) in enumerate(zip(data, categories, colors)):
    ax.bar(x + (i - 2) * width, category_data, width, label=category_label, color=color, alpha=0.9)

# Configurar el gráfico
ax.set_ylabel('Nivel de Impacto Negativo (Daño)', fontsize=12)
ax.set_title('Control de Daños Expandido: Comparativa Metabólica Total', fontsize=16)
ax.set_xticks(x)
ax.set_xticklabels(labels, fontsize=11)
ax.set_ylim(0, 11)  # Asegurar que el 10 sea visible
ax.legend(title='Bebida', bbox_to_anchor=(1.05, 1), loc='upper left')

# Añadir líneas de referencia de "Zona de Seguridad"
ax.axhline(4, color='green', linestyle='--', alpha=0.3)
ax.text(2.6, 3.5, 'Zona de bajo impacto', color='green', fontsize=10, weight='bold')

plt.tight_layout()
plt.savefig('damage_control_expanded_alcohol.png')
```

Aquí tienes la visualización del **Control de Daños Expandido**, comparando cinco escenarios metabólicos distintos.

### Análisis de los Nuevos Escenarios (Orientación Táctica):

El gráfico muestra claramente por qué el acompañante de la bebida es tan crucial como el alcohol en sí:

#### 1. Destilado + Coca-Cola Normal (Barra Roja - Peor Escenario)

* **Carga Glucémica Extrema (10/10):** El alcohol ya inhibe la capacidad del hígado para regular la glucosa. Al añadir el azúcar masivo de una Coca-Cola normal, el cuerpo entra en shock insulínico. Esto provoca el **máximo almacenamiento de grasa** y un crash energético posterior.
* **Deshidratación Severa:** El azúcar y la cafeína de la soda actúan en sinergia con el alcohol para acelerar la diuresis (pérdida de agua). Es la receta para la peor resaca y el mayor daño celular.

#### 2. Destilado + Coca-Cola Zero (Barra Gris - Escenario Intermedio)

* **Mitigación de la Glucosa (1/10):** Eliminamos la carga glucémica por completo. El cuerpo solo tiene que lidiar con el etanol puro, lo cual es un avance enorme respecto a la versión normal.
* **Mantiene Deshidratación e Inflamación:** Aunque es mejor que la versión con azúcar, los edulcorantes artificiales y el ácido fosfórico aún requieren procesamiento hepático y pueden causar micro-inflamación digestiva, afectando la absorción de nutrientes y la recuperación muscular post-entrenamiento.

#### 3. Conclusión de Control de Daños para el Atleta

El **Destilado con Agua (o Agua con Gas)** sigue siendo la "opción de seguridad" (Zona verde). Sin embargo, si el usuario *necesita* sabor, el gráfico le muestra que **la opción Zero es metabólicamente superior a la versión con azúcar**, aunque inferior al agua pura para mantener el estado de hidratación óptimo.

Este gráfico permite a **Vollta** ofrecer una guía de decisión clara: *"Si vas a tomar un Ron, hazlo con Zero para proteger tu metabolismo, o mejor con Agua con Gas para proteger también tu hidratación"*.
